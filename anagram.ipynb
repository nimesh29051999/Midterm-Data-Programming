```python

def freq(w):
    f = [0 for i in range(26)]
    for c in list(w):
        print(ord(c))
        f[ord(c)%97] = 1
    return f
```

```python
def anagram(file, word):
    f = open(file)
    words = []
    res = []
    for line in f.readlines():
        words.append(line.split("\n")[0])
    f1 = freq(word)
    for w in words:
        f2 = freq(w)
        if f1 == f2:
            res.append(w)
    return res
        
```

```python
anagram("words.txt", "trace")
```
